# Mileage Prediction

# Ojective- To predict the mileage by regression analysis of the dataset


**Data** **source** :- This data is taken from the statLib library which id maintained at carnegie Mellon University. This dataset was used in the 1983 American statistical Association Exposition.

**Import** **Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import** **Data**

```python
Mileage=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/MPG.csv')
```

```python
Mileage.head()
```

```python
Mileage.info()
```

```python
Mileage.describe()
```

**Data**:- this data is slightly modified version of data provided in StatLib library. In the line with the use by Ross Quinlan(1993) in predicting the attribute "mpg", 8 of the original instances were removed because they had unknown values for the "mpg" attribute. The original data is available in the file "auto-mpg.data-original"
the data concerns city-cycle fuel consumption in miles per gallon, to be predict three multivalue discrete and 5 continious attributes.

**Data** **Visualiszation**

```python
sns.pairplot( Mileage, x_vars=['displacement','horsepower','weight','acceleration','mpg'], y_vars=['mpg']);
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = Mileage );
```

**Data** **processing**

```python
Mileage.info()
```

```python
Mileage.describe()
```

**Removing** **Missing** **values**

```python
Mileage = Mileage.dropna()
```

```python
Mileage.info()
```

**Define** **Target** **Variable** **y** **and** **feature** **x**

```python
Mileage.columns
```

```python
y = Mileage['mpg']
```

```python
y.shape
```

```python
x = Mileage[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
x.shape
```

```python
x
```

**Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

**Train** **Test** **split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7 , random_state = 2529 )
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**Modeling**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

**Model Evaluation**

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

mileage = 23.48-1.05 displacement = -1.68, horsepower = -4.10, weight = 0.115, acceleration = +error

**Prediction**

```python
y_pred= model.predict(x_test)
```

```python
y_pred
```

**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

Here error percentage is 14, which means accuracy of my model is (100-14)% = 86%
